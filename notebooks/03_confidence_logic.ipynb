```python
# Case 1: diseases are rows
disease_names = disease_symptom_df.iloc[:, 0].tolist()

disease_names
```

```python
symptom_names = disease_symptom_df.columns[1:].tolist()
len(symptom_names), symptom_names[:5]
```

```python
disease_to_symptoms = {}

for _, row in disease_symptom_df.iterrows():
    disease = row.iloc[0]
    relevant_symptoms = []

    for symptom in symptom_names:
        if row[symptom] == 1:
            relevant_symptoms.append(symptom)

    disease_to_symptoms[disease] = relevant_symptoms

disease_to_symptoms
```

```python
symptom_id_to_questions = {}

for _, row in questions_df.iterrows():
    symptom_id = row['symptom_id']
    question_text = row['question']

    if symptom_id not in symptom_id_to_questions:
        symptom_id_to_questions[symptom_id] = []

    symptom_id_to_questions[symptom_id].append(question_text)

list(symptom_id_to_questions.items())[:3]
```

```python
disease_to_symptoms
```

```python
def get_questions_for_disease(disease, max_questions=3):
    symptom_ids = disease_to_symptoms.get(disease, [])
    selected_questions = []

    for sid in symptom_ids:
        if sid in symptom_id_to_questions:
            selected_questions.extend(symptom_id_to_questions[sid])

        if len(selected_questions) >= max_questions:
            break

    return selected_questions[:max_questions]
```

```python
test_disease = 'mel'   # example
questions = get_questions_for_disease(test_disease)

print("Disease:", test_disease)
for q in questions:
    print("-", q)
```

```python
ANSWER_SPACE = ["YES", "NO", "UNKNOWN"]
```

```python
def initialize_answer_store(questions):
    return {q: None for q in questions}
```

```python
answer_store = initialize_answer_store(questions)
answer_store
```

```python
# Simulated user responses (temporary)
answer_store['Has the lesion changed in color?'] = "YES"
answer_store['Is the lesion asymmetrical?'] = "YES"
answer_store['Are the borders irregular or uneven?'] = "NO"

answer_store
```

```python
question_to_symptom_id = {}

for _, row in questions_df.iterrows():
    question_to_symptom_id[row['question']] = row['symptom_id']

list(question_to_symptom_id.items())[:3]
```

```python
def build_symptom_evidence(answer_store):
    symptom_evidence = {}

    for question, answer in answer_store.items():
        symptom_id = question_to_symptom_id.get(question)

        if symptom_id is None:
            continue

        symptom_evidence[symptom_id] = answer

    return symptom_evidence
```

```python
symptom_evidence = build_symptom_evidence(answer_store)
symptom_evidence
```

```python
SYMPTOM_ANSWER_WEIGHT = {
    "YES": 1.0,
    "NO": -0.5,
    "UNKNOWN": 0.0
}
```

```python
def compute_symptom_support(symptom_evidence):
    score = 0.0
    for answer in symptom_evidence.values():
        score += SYMPTOM_ANSWER_WEIGHT.get(answer, 0.0)
    return score
```

```python
symptom_support_score = compute_symptom_support(symptom_evidence)
symptom_support_score
```

```python
def normalize_symptom_score(score, max_possible=3.0):
    normalized = score / max_possible
    return max(-1.0, min(1.0, normalized))
```

```python
def fuse_confidence(image_confidence, symptom_score, alpha=0.7):
    symptom_signal = normalize_symptom_score(symptom_score)
    fused = (alpha * image_confidence) + ((1 - alpha) * (image_confidence + symptom_signal * 0.2))
    return max(0.0, min(1.0, fused))
```

```python
image_confidence = 0.31  # simulated CNN confidence

fused_confidence = fuse_confidence(
    image_confidence,
    symptom_support_score
)

print("Image confidence :", round(image_confidence, 2))
print("Symptom score    :", symptom_support_score)
print("Fused confidence :", round(fused_confidence, 2))
```

```python
HIGH_CONF_THRESHOLD = 0.70
MID_CONF_THRESHOLD  = 0.40
LOW_CONF_THRESHOLD  = 0.25
```

```python
def decide_action(confidence):
    if confidence >= HIGH_CONF_THRESHOLD:
        return "SAFE_ADVISORY"
    elif confidence >= MID_CONF_THRESHOLD:
        return "ASK_MORE_QUESTIONS"
    elif confidence >= LOW_CONF_THRESHOLD:
        return "LOW_CONFIDENCE_WARNING"
    else:
        return "ETHICAL_REFUSAL"
```

```python
initial_action = decide_action(image_confidence)
final_action   = decide_action(fused_confidence)

print("Initial image confidence :", round(image_confidence, 2))
print("Initial decision         :", initial_action)
print("Fused confidence         :", round(fused_confidence, 2))
print("Final decision           :", final_action)
```
